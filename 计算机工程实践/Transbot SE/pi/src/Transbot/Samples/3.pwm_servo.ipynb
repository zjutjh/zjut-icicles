```python
#!/usr/bin/env python3
#coding=utf-8
import time
from Transbot_Lib import Transbot
from ipywidgets import interact
import ipywidgets as widgets
```

```python
# 创建Transbot对象 bot
# Create Transbot object as bot
bot = Transbot()
```

滑块widgets.IntSlider参数解释：
min=最小值,max=最大值,step=最小变化量,value=默认值
interact将滑块显示出来，并且负责处理滑动产生的事件。  

Slider widgets.IntSlider parameter  
Min = minimum, Max = maximum,step= minimum change,value= default  
Interact displays the slider and is responsible for handling the events generated by the slide.  

```python
# 深度摄像头控制舵机云台。增加最大值和最小值限制
# Depth camera controls steering gear head.  Increase the maximum and minimum limits  
def pwm_servo_x(x):
    bot.set_pwm_servo(1, x)
    return x

# 创建一个滑块来控制舵机左右转动
# Create a slider to control the steering gear to turn left and right
interact(pwm_servo_x, x=widgets.IntSlider(min=30,max=150,step=1,value=90));
```

```python
# 普通摄像头的二自由度云台控制。
# 注意：深度相机不要运行此命令，由于没有限制角度可能会撞上其他东西。
# General camera two degrees of freedom control head.
# Note: Do not run this command for Astra camera, as there is no limit to the Angle you may bump into something else.  
def pwm_servo(x, y):
    bot.set_pwm_servo(1, x)
    bot.set_pwm_servo(2, y)
    return x, y

# 创建两个滑块来控制舵机云台
# Create two sliders to control the steering head
interact(pwm_servo, x=widgets.IntSlider(min=0,max=180,step=1,value=90), \
         y=widgets.IntSlider(min=0,max=180,step=1,value=90));
```

```python
# 程序结束后请删除transbot对象，避免在其他程序中使用transbot库造成冲突
# After the program is complete, delete the Transbot object to avoid conflicts caused by using the Transbot library in other programs
del bot
```
